## Задача 1

В партии 10 деталей.
<br>
Среди них 3 бракованные.
<br>
Какова вероятность, что среди 5 ,взятых на удачу, 4 хорошие детали?

##### Решение:

```python
n = 10 # Количество деталей всего
b = 3  # Количество бракованных деталей
k = 5  # Количество выбираемых случайным образом деталей


# Всего способов выбрать k из n без учета порядка - С(k, n)
total_combinations = math.comb(n, k)
print(f'Всего способов выбрать {k} деталей из {n}: {total_combinations}')

# 'Благоприятные' комбинации - те, в которых ровно 4 хороших детали и одна бракованная.
#  Число комбинаций с 4 хорощими деталями - C(n-b, 4)  - число способов выбрать 4 элемента из 7 без учета порядка
total_good = math.comb(n-b, k-1)
print(f'Всего способов выбрать {k-1} хороших деталей из {n-b}: {total_good}')
#  Число комбинаций с 1 бракованной деталью - C(b, 1)  - число способов выбрать 1 элемента из 3 без учета порядка (3 способа)
total_bad = math.comb(b, 1)
print(f'Всего способов выбрать 1 бракованную деталь из {b}: {total_bad}')
# 'Благоприятные' комбинации
total_l = total_bad*total_good
print(f'Всего способов выбрать 1 бракованную деталь и {k-1} хороших из {n}: {total_bad}*{total_good}={total_l}')

# Вероятность
p = total_l / total_combinations
print(f'Вероятность, что среди 5 деталей, взятых на удачу из 10 (7 хороших и 3 бракованных) будут 4 хорошие детали, составляет {p*100:.3f}%')
```

## Задача 2

Разработали спам-фильтр на основании часто встречающихся фраз. 70% всех писем –это
<br>
спам. В 10 % писем со спамом встречается фраза: «вся правда о» и в 0.5% она встречается
<br>
в хороших письмах. Какова вероятность, что пришедшее на почту письмо является спамом,
<br>
если в нем есть данная фраза?

##### Решение:

```python
# Среди 10000 писем 7000 СПАМ и 3000 НЕ СПАМ, при этом 7000*0.1 = 700 писем - СПАМ с заданой фразой и 3000*0.005 = 3*5 = 15 - НЕ СПАМ с заданной фразой, всего писем с заданной фразой
#  - 715 писем.
# Тогда число благоприятных исходов - 700 (СПАМ с заданой фразой), а полное число всех исходов - 715 (число всех писем среди 10000 с заданной фразой)
# Считаем вероятность:
p = 700/715
print(f'Вероятность того, что пришедшее письмо является спамом, если в нем есть данная фраза {p*100:.3f}%')

# Или по формуле Байеса:
# P(B|A) = (P(A/B)*P(B))/P(A), где
# A - письмо содержит заданную фразу
# B - письмо является спамом
# P(B) = 0.7 по условию
# P(A|B) = 0.1  по условию
# Cобытие A является суммой несовместных событий А1 и А2, где А1 - письмо является спамом и содержит заданную фразу, а А2 - письмо не является спамом и содержит заданную фразу
# поэтому P(A) = P(A1)+P(A2),
# при этом событие А1 - произведение двух событий - 'письмо является спамом' и 'письмо содержит заданную фразу',
# поэтому P(А1) = P(В)*P(A|B) = 0.7*0.1.
# Аналогично, P(A2) = 0.3*0.005 и P(A) = 0.7*0.1+0.3*0.005 ==>
# P(B|A) = (0.1*0.7)/(0.7*0.1+0.3*0.005)
print(f"Вероятность того, что пришедшее письмо является спамом, если в нем есть данная фраза (по формуле Байеса): {0.7*0.1*100/(0.1*0.7 + 0.3*0.005):.3f}%")
```

## Задача 3

В ящике находится 10 красных, 5 черных, 5 зеленых шаров. Наудачу вынимают 6 шаров.
<br>
Какова вероятность, что вынуты 3 красных, 2 черных, 1 зеленый?

##### Решение:

```python
# Фактически задача сводится к определению того, какова доля требуемых выборок среди всех выборок из 6 шаров.
# Всего выборок из 6 шаров без учета порядка - С(6, total_balls)
total_balls = 10 + 5 + 5
total_combinations = math.comb(total_balls, 6)

# Нас интересуют выборки из 3 красных, 2 черных и 1 зеленого без учета порядка:
# С(3, 10) - количество выборок 3 красных из 10 красных
# С(2, 5) - количество выборок 2 черных из 5 черных
# С(1, 5) - количество выборок 1 зеленого из 5 зеленых, тут интуитивно ясно, что это 5
# Так как цвета тут независимы, то число интересующих нас выборок можно получить произведением:
# Сначала выбрали все комбинации 3 красных из 10, затем для каждой выбрали 2 черных из 5 и наконец, затем 1 зеленый из 5.
combinations = math.comb(10, 3)*math.comb(5, 2)*math.comb(5, 1)

print(f'Вероятность составляет {(combinations/total_combinations)*100:3f}%')
```

## Задача 4

На 5 одинаковых карточках написаны буквы Ч, А, Й ,К, И
<br>
Какова вероятность, что получится слова ЧАЙКИ

##### Решение:

```python
# Число всех возможных комбинаций - число перестановок из 5 элементов (число слов из 5 заданных букв).
total_combinations = math.factorial(5)

# Так как все буквы разные, то количество требуемых комбинаций - ровно 1 - слово "Чайки" можно из заданных букв получить только одной расстановой.
combinations = 1

print(f'Вероятность с первого раза из указанных карточек составить требуемое слово: {(combinations / total_combinations)*100:.3f}%')
```

## Задача 6

На 5 одинаковых карточках написаны буквы Ч, А, Й ,К, А.
<br>
Какова вероятность, что получится слова ЧАЙКА

##### Решение:

```python
# Число всех возможных комбинаций - число перестановок из 5 элементов (число слов из 5 заданных букв).
total_combinations = math.factorial(5)

# Так как не все буквы разные, то количество требуемых комбинаций - ровно 2 - слово "Чайка" можно из заданных букв получить 2 -мя расстановками
# (когда на втором месте вторая карточка, а на 5-ом 5-ая или когда на втором - пятая, а 5-ом - вторая).
combinations = 2

# Можно найти это число и так:
# Способов выбрать первой букву Ч - 1
# Способов выбрать второй букву А - 2
# Способов выбрать третьей букву Й - 1
# Способов выбрать четверой букву К - 1
# Способов выбрать пятой букву А - 1 (другая буква А уже выбрана).
# Итого: 1*2*1*1*1 = 2

print(f'Вероятность с первого раза из указанных карточек составить требуемое слово: {(combinations / total_combinations)*100:.3f}%')
```
